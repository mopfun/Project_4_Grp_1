```python
# Import our dependencies
import pandas as pd
from pathlib import Path
```

```python
path = Path("Resources/final_combined_data.csv")
df = pd.read_csv(path)
df.head()
```

```python
# df['cast0']
```

```python
cast_df = df[['cast1', 'cast2', 'cast3', 'cast4']]
cast_df
```

```python
# Count duplicates per row across columns
cast_df['duplicate_count'] = cast_df.apply(lambda row: len(row) - len(set(row)), axis=1)
# print(cast_df)
cast_df.head()
```

```python
#alt option to see if there are any duplicates by 'flagging' it with a true or false stmt
cast_df['has_duplicate'] = cast_df.apply(lambda row: len(set(row)) < len(row), axis=1)
# print(df)
cast_df.head()
```

```python
#showing the dup counts in cast1 row
cast1_counts = cast_df['cast1'].value_counts()
# cast1_counts
cast2_counts = cast_df['cast2'].value_counts()
# cast2_counts
cast3_counts = cast_df['cast3'].value_counts()
# cast3_counts
cast4_counts = cast_df['cast4'].value_counts()
# cast4_counts

print(cast1_counts)
print(cast2_counts)
print(cast3_counts)
print(cast4_counts)
```
