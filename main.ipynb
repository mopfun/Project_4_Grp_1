```python
imdb_df.rename(columns={"Duration": "Prime_Duration"}, inplace=True)
imdb_df.head()
```

```python
# convert the duration column into decimal format
def duration_to_decimal(x):
    try:
        # Default values
        hours = 0
        minutes = 0

        # Only parse if value is a string
        if isinstance(x, str):
            if 'h' in x:
                hours = int(x.split('h')[0].strip())
            if 'm' in x:
                minutes = int(x.split('h')[1].split('m')[0].strip()) if 'h' in x else int(x.split('m')[0].strip())

        return round(hours + minutes / 60, 4)
    except Exception as e:
        return None  # Or np.nan

imdb_df['Duration'] = imdb_df['Prime_Duration'].apply(duration_to_decimal)
imdb_df.head()

# save df into a csv fiile
# imdb_df.to_csv('updated_imdb_data.csv', index=False)
```

```python
# merging the two datasets: imdb and netflix filtered wkly global
# Merge DataFrames
merged_df = pd.merge(wkly_global, imdb_df, on='Duration', how='outer')
merged_df.head()

#save df
# merged_df.to_csv('merged_data.csv', index=False)
```
