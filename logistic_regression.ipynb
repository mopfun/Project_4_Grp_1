```python
# Import our dependencies
import pandas as pd
from pathlib import Path
# from sklearn.model_selection import train_test_split
# from sklearn.preprocessing import StandardScaler
# import tensorflow as tf
```

```python
path = Path("Resources/final_combined_data.csv")
df = pd.read_csv(path)
df.head()
```

```python
filtered_df = df[['row_id', 'rating', 'votes', 'meta_score', 'number_rating', 'weekly_rank', 'weekly_hours_viewed', 'cumulative_weeks_in_top_10' ]]
filtered_df.head()
```

### Designating target and features

```python
#target (y variable)
y = filtered_df['cumulative_weeks_in_top_10']

#feature (X variable)
X = filtered_df.drop(columns=['cumulative_weeks_in_top_10'])
```

```python
#review y variable series
y.head()
```

```python
#review X variable DF
X.head()
```

### Split the data into training and testing datasets by using `train_test_split`.

```python
# Import the train_test_learn module
from sklearn.model_selection import train_test_split
```

```python
#split data and use randfom_state of 1
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

### Create logistic model with original data

```python
# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression
```

```python
# instantiate logistical regression model with random_state of 1
model = LogisticRegression(random_state=1)

#fit model using training data
model.fit(X_train, y_train)
```

### Save predictions of test data labels using testing featire (X_test)

```python
# make prediction
y_pred = model.predict(X_test)
```

### Evaluate model's performance by: generating confusion matrix and printing classification report

```python
# import dependencies
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Generate a confusion matrix for the model
matrix = confusion_matrix(y_test, y_pred)

print(matrix)
```

```python
# Print the classification report for the model
classification_report = classification_report(y_test, y_pred)

print(classification_report)
```

```python
# Confusion matrix
matrix = confusion_matrix(y_test, y_pred)

# Get number of classes
num_classes = matrix.shape[0]

# Dynamically create labels
index_labels = [f"Actual {i}" for i in range(num_classes)]
column_labels = [f"Predicted {i}" for i in range(num_classes)]

# Create DataFrame
matrix_df = pd.DataFrame(matrix, index=index_labels, columns=column_labels)

# Display
matrix_df.head()
# print(matrix_df)
```

```python
from sklearn.metrics import accuracy_score

# Display the accuracy score for the test dataset.
acc_score = accuracy_score(y_test, y_pred)
acc_score
```
